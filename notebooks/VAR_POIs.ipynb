```python
import pandas as pd
import geopandas as gpd
import os
```

```python
raw_path = os.path.join('..', 'data', 'raw')
```

```python
areas_freguesias = {
    'Ajuda': 2.88,
    'Alcântara': 5.07,
    'Alvalade': 5.34,
    'Areeiro': 1.74,
    'Arroios': 2.13,
    'Avenidas Novas': 2.99,
    'Beato': 2.48,
    'Belém': 10.43,
    'Benfica': 8.02,
    'Campo de Ourique': 1.65,
    'Campolide': 2.77,
    'Carnide': 3.69,
    'Estrela': 4.60,
    'Lumiar': 6.57,
    'Marvila': 7.12,
    'Misericórdia': 2.19,
    'Olivais': 8.09,
    'Parque das Nações': 5.44,
    'Penha de França': 2.71,
    'Santa Clara': 3.36,
    'Santa Maria Maior': 3.01,
    'Santo António': 1.49,
    'São Domingos de Benfica': 4.29,
    'São Vicente': 1.99
}
```

```python
shp_metro = gpd.read_file(os.path.join(raw_path, 'estacoes_metro', 'Estações_de_Metro.shp'))
shp_comboio = gpd.read_file(os.path.join(raw_path, 'estacoes_comboio', 'Estações_de_Comboio.shp'))
shp_saude = gpd.read_file(os.path.join(raw_path, 'prestacao_servicos_saude', 'Prestação_de_Cuidados.shp'))
```

```python
shp_metro.head()
```

```python
shp_comboio
```

```python
caop = gpd.read_file(os.path.join(raw_path, 'Continente_CAOP2024_1.gpkg'))
```

```python
caop.head()
```

```python
metro = shp_metro.to_crs(caop.crs)
comboio = shp_comboio.to_crs(caop.crs)
saude = shp_saude.to_crs(caop.crs)
```

```python
print(metro.crs)
print(comboio.crs)
print(saude.crs)
print(caop.crs)
```

```python
metro_freg = gpd.sjoin(metro, caop, how="inner", predicate='within')
comboio_freg = gpd.sjoin(comboio, caop, how="inner", predicate='within')
saude_freg = gpd.sjoin(saude, caop, how="inner", predicate='within')
```

```python
# contagem estações de metro por freguesia de lisboa (incluir aquelas que têm 0 estações)
contagem_metro = (
    metro_freg
    .groupby('freguesia')
    .size()
    .reset_index(name='Num_Estacoes_Metro')
)

contagem_comboio = (
    comboio_freg
    .groupby('freguesia')
    .size()
    .reset_index(name='Num_Estacoes_Comboio')
)

contagem_saude = (
    saude_freg
    .groupby('freguesia')
    .size()
    .reset_index(name='Num_Prestadores_Saude')
)
```

```python
contagem_metro
```

```python
contagem_comboio
```

```python
metro_freg = gpd.sjoin(
    metro,
    caop,
    how='left',
    predicate='within'
)

comboio_freg = gpd.sjoin(
    comboio,
    caop,
    how='left',
    predicate='within'
)

saude_freg = gpd.sjoin(
    saude,
    caop,
    how='left',
    predicate='within'
)
```

```python
comboio_freg
```

```python
# contagem estações de metro e comboio por freguesia de lisboa (incluir aquelas que têm 0 estações)
contagem_metro = (
    metro_freg
    .groupby('freguesia')
    .size()
    .reset_index(name='Num_Estacoes_Metro')
)

comboio_freg = (
    comboio_freg
    .groupby('freguesia')
    .size()
    .reset_index(name='Num_Estacoes_Comboio')
)

saude_freg = (
    saude_freg
    .groupby('freguesia')
    .size()
    .reset_index(name='Num_Prestadores_Saude')
)
```

```python
# Merge com freguesias
freguesias = caop.merge(
    contagem_metro,
    on="freguesia",
    how="left"
)

freguesias = freguesias.merge(
    contagem_comboio,
    on="freguesia",
    how="left"
)

freguesias = freguesias.merge(
    contagem_saude,
    on="freguesia",
    how="left"
)

# Substituir NA por 0 (freguesias sem metro)
freguesias["Num_Estacoes_Metro"] = freguesias["Num_Estacoes_Metro"].fillna(0).astype(int)
freguesias["Num_Estacoes_Comboio"] = freguesias["Num_Estacoes_Comboio"].fillna(0).astype(int)
freguesias["Num_Prestadores_Saude"] = freguesias["Num_Prestadores_Saude"].fillna(0).astype(int)
```

```python
#freguesias municipio de lisboa
freguesias_lisboa = freguesias[freguesias['municipio'] == 'Lisboa']
freguesias_lisboa.head()
```

```python
## deixar só as colunas das freguesias e Num_
freguesias_lisboa = freguesias_lisboa[['freguesia', 'Num_Estacoes_Metro', 'Num_Estacoes_Comboio', 'Num_Prestadores_Saude']]
freguesias_lisboa['Area_km2'] = freguesias_lisboa['freguesia'].map(areas_freguesias)
freguesias_lisboa
```

```python
# desnsidade de estações de metro por km2
freguesias_lisboa['Densidade_Estacoes_Metro_km2'] = freguesias_lisboa['Num_Estacoes_Metro'] / freguesias_lisboa['Area_km2']
freguesias_lisboa['Densidade_Estacoes_Comboio_km2'] = freguesias_lisboa['Num_Estacoes_Comboio'] / freguesias_lisboa['Area_km2']
freguesias_lisboa['Densidade_Prestadores_Saude_km2'] = freguesias_lisboa['Num_Prestadores_Saude'] / freguesias_lisboa['Area_km2']
freguesias_lisboa
```

```python
path_ensino = os.path.join(raw_path, 'ensino_superior', 'ensino_superior.csv')
ensino_superior = pd.read_csv(path_ensino)

ensino_superior_freg = (
    ensino_superior
    .groupby('Freguesia')
    .size()
    .reset_index(name='Num_Estabelecimentos_Ensino_Superior')
)
ensino_superior_freg

# merge com freguesias_lisboa sem adiconar coluna Freguesia duas vezes
freguesias_lisboa = freguesias_lisboa.merge(
    ensino_superior_freg,
    left_on='freguesia',
    right_on='Freguesia',
    how='left'
).drop(columns=['Freguesia'])

freguesias_lisboa['Num_Estabelecimentos_Ensino_Superior'] = freguesias_lisboa['Num_Estabelecimentos_Ensino_Superior'].fillna(0).astype(int)
freguesias_lisboa['Densidade_Estabelecimentos_Ensino_Superior_km2'] = freguesias_lisboa['Num_Estabelecimentos_Ensino_Superior'] / freguesias_lisboa['Area_km2']
freguesias_lisboa
```

```python
### Escolas Públicas
public_schools_dir = os.path.join(raw_path, 'escolas_publicas')

escolas_pre_escolar = pd.read_csv(os.path.join(public_schools_dir, 'escola_publica_pre_escolar.csv'))
escolas_1_ciclo = pd.read_csv(os.path.join(public_schools_dir, 'escola_publica_1ciclo.csv'))
escolas_2_3_ciclo = pd.read_csv(os.path.join(public_schools_dir, 'escola_publica_2_3ciclo.csv'))
escolas_3_ciclo_secundario = pd.read_csv(os.path.join(public_schools_dir, 'escola_publica_3ciclo_secundario.csv'))
escolas_secundario = pd.read_csv(os.path.join(public_schools_dir, 'escola_publica_secundario.csv'))
escolas_publicas = pd.concat([
    escolas_pre_escolar,
    escolas_1_ciclo,
    escolas_2_3_ciclo,
    escolas_3_ciclo_secundario,
    escolas_secundario
])
escolas_publicas.head()
```

```python
## verificar duplicados (nomes das escolas)
escolas_publicas.duplicated().sum()
```

```python
## tratar dos duplicados (remover duplicados, só deixar uma ocorrência)
escolas_publicas = escolas_publicas.drop_duplicates()
escolas_publicas = escolas_publicas.reset_index(drop=True)
escolas_publicas.head()
```

```python
# Num_Escolas_Publicas em freguesias de lisboa
escolas_publicas_freg = (
    escolas_publicas
    .groupby('Freguesia')
    .size()
    .reset_index(name='Num_Escolas_Publicas')
)
escolas_publicas_freg
```

```python
## merge com freguesias_lisboa sem adiconar coluna Freguesia duas vezes
freguesias_lisboa = freguesias_lisboa.merge(
    escolas_publicas_freg,
    left_on='freguesia',
    right_on='Freguesia',
    how='left'
).drop(columns=['Freguesia'])
freguesias_lisboa['Num_Escolas_Publicas'] = freguesias_lisboa['Num_Escolas_Publicas'].fillna(0).astype(int)

# fazer densidade de escolas públicas por km2
freguesias_lisboa['Densidade_Escolas_Publicas_km2'] = freguesias_lisboa['Num_Escolas_Publicas'] / freguesias_lisboa['Area_km2']
freguesias_lisboa
```

```python
### Escolas Privadas
private_schools_dir = os.path.join(raw_path, 'escolas_privadas')
escolas_pre_escolar = pd.read_csv(os.path.join(private_schools_dir, 'escola_privada_pre_escolar.csv'))
escolas_1_ciclo = pd.read_csv(os.path.join(private_schools_dir, 'escola_privada_1ciclo.csv'))
escolas_2_3_ciclo = pd.read_csv(os.path.join(private_schools_dir, 'escola_privada_2_3ciclo.csv'))
escolas_secundario = pd.read_csv(os.path.join(private_schools_dir, 'escola_privada_secundario.csv'))
escolas_privadas = pd.concat([
    escolas_pre_escolar,
    escolas_1_ciclo,
    escolas_2_3_ciclo,
    escolas_secundario
])
escolas_privadas.head()
```

```python
## verificar duplicados (nomes das escolas)
escolas_privadas.duplicated().sum()
```

```python
# Num_Escolas_Privadas em freguesias de lisboa
escolas_privadas_freg = (
    escolas_privadas
    .groupby('Freguesia')
    .size()
    .reset_index(name='Num_Escolas_Privadas')
)
escolas_privadas_freg
```

```python
## merge com freguesias_lisboa sem adiconar coluna Freguesia duas vezes
freguesias_lisboa = freguesias_lisboa.merge(
    escolas_privadas_freg,
    left_on='freguesia',
    right_on='Freguesia',
    how='left'
).drop(columns=['Freguesia'])
freguesias_lisboa['Num_Escolas_Privadas'] = freguesias_lisboa['Num_Escolas_Privadas'].fillna(0).astype(int)

# fazer densidade de escolas privadas por km2
freguesias_lisboa['Densidade_Escolas_Privadas_km2'] = freguesias_lisboa['Num_Escolas_Privadas'] / freguesias_lisboa['Area_km2']
freguesias_lisboa
```
